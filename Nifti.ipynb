```python
image_data= nifti_file.get_fdata()
```

```python
fig , axis = plt.subplots(3,3,figsize = (15,15))

slice_counter = 0

for i in range(3):
  for j in range(3):
    axis[i][j].imshow(image_data[:,:,slice_counter], cmap = "gray")
    slice_counter +=1
```

```python
image_data_processed = image_data * (image_data > 250)

fig , axis = plt.subplots(3,3,figsize = (15,15))

slice_counter = 0

for i in range(3):
  for j in range(3):
    axis[i][j].imshow(image_data_processed[:,:,slice_counter], cmap = "gray")
    slice_counter +=1
```

```python
processed_nifti = nib.Nifti1Image(image_data_processed, nifti_file.affine )
```

```python
nib.save(processed_nifti , "drive/MyDrive/Nifti/test.nii.gz")
```
